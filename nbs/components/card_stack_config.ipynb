## Configuration

The alignment card stack uses a unique `sd-align` prefix to avoid ID collisions with the
decomposition card stack (`sd-decomp`). Cards are more compact (time ranges vs. text content),
so defaults use narrower width range and tighter spacing.

```python
#| export
ALIGN_CS_CONFIG = CardStackConfig(
    prefix="sd-align",
    card_width_min=20,       # Narrower range for compact cards
    card_width_max=60,
    card_width_step=5,
    style=CardStackStyleConfig(
        section_gap="0.75rem",     # Tighter spacing for compact cards
        slot_padding="0.125rem",
        viewport_padding_x='0.0rem', 
        viewport_padding_y='0.0rem'
    ),
)

ALIGN_CS_IDS = CardStackHtmlIds(prefix=ALIGN_CS_CONFIG.prefix)
ALIGN_CS_BTN_IDS = CardStackButtonIds(prefix=ALIGN_CS_CONFIG.prefix)
```
